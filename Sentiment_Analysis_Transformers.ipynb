```python
# ËºâÂÖ•Áõ∏ÈóúÂ•ó‰ª∂
from transformers import pipeline
```

```python
# ËºâÂÖ•Ê®°Âûã
classifier = pipeline('sentiment-analysis')
```

```python
# Ê≠£Èù¢
print(classifier('We are very happy to show you the ü§ó Transformers library.'))

# Ë≤†Èù¢
print(classifier('I hate this movie.'))

# Âê¶ÂÆöÂè•‰πüÂèØ‰ª•Ê≠£Á¢∫ÂàÜÈ°û
print(classifier('the movie is not bad.'))
```

```python
# Ê∏¨Ë©¶Â§öÁ≠Ü
results = classifier(["The plot is quite simple, the theme is great though.", 
                      "The plot is excellent, but the characterization lets me down."])
for result in results:
    print(f"label: {result['label']}, with score: {round(result['score'], 4)}")
```
